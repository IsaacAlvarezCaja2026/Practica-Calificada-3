```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://books.toscrape.com/"
respuesta = requests.get(url)
respuesta.encoding = "utf-8"  # Corregido de "uft-8" a "utf-8"
soup = BeautifulSoup(respuesta.text, "html.parser")
libros = soup.select("article", class_="product_pod")

libros_lista = []

for i, libro in enumerate(libros[:20], start=1):
    titulo = libro.h3.a["title"]
    precio = libro.find("p", class_="price_color").text

    libros_lista.append({"Título": titulo, "Precio": precio})

df = pd.DataFrame(libros_lista, columns=["Título", "Precio"])
df.insert(0, "N°", range(1, len(df) + 1))
print(df)
df.to_csv("libros.csv", index=False, encoding="utf-8-sig")

# --- LÍNEA CORREGIDA AQUÍ ---
df["Precio"] = df["Precio"].str.replace("£", "", regex=False).astype(float)

libro_caro = df.loc[df["Precio"].idxmax()]
libro_barato = df.loc[df["Precio"].idxmin()]

print("\nLibro más caro de la lista:")
print(libro_caro)

print("\nLibro más barato de la lista:")
print(libro_barato)

```
